## 2) Guide to start a Machine Learning Project 

#### There are eight main steps:

1. Frame the problem and look at the big picture.
2. Get the data.
3. Explore the data to gain insights.
4. Prepare the data to better expose the underlying data patterns to Machine Learning algorithms.
5. Explore many different models and short-list the best ones.
6. Fine-tune your models and combine them into a great solution.
7. Present your solution.
8. Launch, monitor, and maintain your system.

<u>Obviously, you should feel free to adapt this checklist to your needs.</u>

### Performance Measure

- A typical performance measure for regression problems in **Root Mean Squared Error (RMSE)**


$$
RMSE(X,y,h) =
\sqrt{
\frac{1}{m}
\sum_{i=1}^{m}
\left(h(x^{(i)}) - y^{(i)}\right)^2
}
$$

- **Notations:**
    1) *m* is the number of instances in dataset.
        - For example, if you are evaluating the RMSE on a validation set of 2000 districts, then *m* = 2000
    2) *$x^{(i)}$* is the vector of all the feature values (excluding lables) of the $i^{(th)}$ instance in the dataset.
    3) *$y^{(i)}$* is its label (the desired output value of for the instance).
    4) *y* is a vector containing the label of all the instances in the dataset.
        - For example, if the first district dataset is located at longtitude -118.29° latitude 33.91°, and its inhabitants with a median income of $38,372 and the median house value is $156,400 then:

        $$
        x^{(1)} =
        \begin{pmatrix}
        -118.29 \\
        33.91 \\
        1,416 \\
        38,372
        \end{pmatrix}
        $$

        <center>and:</center>

        $$
        y^{(1)} = 156,400
        $$

    5) X is a matrix containg all the feature values
        - For example, if the first disctrict is as just described then the matrix X will look like this:

        $$
        X =
        \begin{pmatrix}
        (x^{(1)})^T \\
        (x^{(2)})^T \\
        \vdots \\
        (x^{(2000)})^T
        \end{pmatrix}
        =
        \begin{pmatrix}
        -118.29 & 33.91 & 1416 & 38372 \\
        -117.85 & 34.12 & 1250 & 29400 \\
        \vdots & \vdots & \vdots & \vdots \\
        -121.45 & 37.77 & 1800 & 45000
        \end{pmatrix}
        $$

    6) *h* is the system's prediction function also called a ***hypothesis***.
